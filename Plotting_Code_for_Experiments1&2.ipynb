```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

spreadsheets = ['Experiment 2']

# List to store standard deviation values
std_dev_values = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.5]

# List of dataframes
dfs = ['0', '1', '2', '3', '4', '5', '6', '7']

for spreadsheet in spreadsheets:
    # Load spreadsheet
    xl = pd.ExcelFile(f'{spreadsheet}.xlsx')

    # Lists to store average similarities and similarities_squared
    avg_similarities = []
    avg_similarities_squared = []

    # Load each sheet into a DataFrame, calculate average similarities and similarities_squared
    for df in dfs:
        data = xl.parse(df)
        data['similarities'] = data['similarities'].apply(ast.literal_eval).str[0]
        data['similarities_squared'] = data['similarities_squared'].apply(ast.literal_eval).str[0]
        avg_similarity = data['similarities'].mean()  # calculate average similarity
        avg_similarity_squared = data['similarities_squared'].mean()  # calculate average similarity squared
        avg_similarities.append(avg_similarity)
        avg_similarities_squared.append(avg_similarity_squared)

    # Plot average similarities against standard deviation values
    plt.figure(figsize=(4, 4))
    plt.plot(std_dev_values, avg_similarities, marker='o')
    plt.xlabel('Standard Deviation (Noise)')
    plt.ylabel('Average Similarity')
    plt.title(f'Average Similarity vs Standard Deviation for {spreadsheet}')
    plt.grid(True)
    plt.show()

    # Plot average similarities_squared against standard deviation values
    plt.figure(figsize=(4, 4))
    plt.plot(std_dev_values, avg_similarities_squared, marker='o')
    plt.xlabel('Standard Deviation (Noise)')
    plt.ylabel('Average Similarity Squared')
    plt.title(f'Average Similarity Squared vs Standard Deviation for {spreadsheet}')
    plt.grid(True)
    plt.show()
```
